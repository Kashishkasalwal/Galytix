```python
array=df_realgdp['Country']
```

```python
df1.head()
```

```python
df_air=df1[df1['lob']=='aviation']
```

```python
df_marine=df1[df1['lob']=='marine']
```

```python
year=np.array([2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015])
```

```python
penetration_air=[];
y=0
for yr in year:
    penet=0
    x=0
    for cotr in array:
        x+=1
        z=(df_air[(df_air['Country']==cotr)][yr])/(df_realgdp[df_realgdp['Country']==cotr][yr])
        if z.notnull:
            penet+=x
    agg_penet=penet/x
    penetration_air.append(agg_penet)
```

```python
penetration_marine=[];
y=0
for yr in year:
    penet=0
    x=0
    for cotr in array:
        x+=1
        z=(df_marine[(df_marine['Country']==cotr)][yr])/(df_realgdp[df_realgdp['Country']==cotr][yr])
        if z.notnull:
            penet+=x
    agg_penet=penet/x
    penetration_marine.append(agg_penet)
```
